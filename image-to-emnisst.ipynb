# Image to EMNIST
Preprocess any image to fit the input of the classifer 

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image, ImageFilter

def image_to_emnist(img):
    img = img.resize((28, 28), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
    
    img_canvas = Image.new('L', (28, 28), (255))
    img_canvas.paste(img)
    
    img_data = list(img_canvas.getdata())

    #Normalize pixels to 0 and 1. 0 is pure white, 1 is pure black.
    img_normalized = [(255 - x) * 1.0 / 255.0 for x in img_data]
    
    return img_normalized
```

## Samples

### Paint image

#### Original

```python
imgPath = './sample_images/m.png'

img = Image.open(imgPath)

plt.imshow(img, cmap='Greys_r')
print('Size', img.size)
```

#### EMNIST ready

```python
img_emnist = np.array(image_to_emnist(img))
img_emnist.shape
```

```python
img_emnist = np.array(img_emnist).reshape(28,28)
img_emnist.shape
```

```python
plt.imshow(img_emnist, cmap='Greys_r')
print('Size:', img_emnist.shape)
```
